```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer
```

```python
#   x_train, y_train
X_train = data.drop(target_col, axis=1)
y_train = data[target_col]
X_test = data_test

#      
X_train_t = X_train.copy()
X_test_t = X_test.copy()
```

```python
data.info()
```

```python
#    
robustscaler = RobustScaler()
standardscaler = StandardScaler()
n_cols = ["TotalSpent"]
num_cols = [col for col in X_train.columns if data[col].dtype != 'object' and col != "TotalSpent"]


X_train_t[n_cols] = robustscaler.fit_transform(X_train_t[n_cols])
X_test_t[n_cols] = robustscaler.transform(X_test_t[n_cols])


X_train_t[num_cols] = standardscaler.fit_transform(X_train_t[num_cols])
X_test_t[num_cols] = standardscaler.transform(X_test_t[num_cols])

X_train_t
```

```python
#     ,    internetservice,   ,        no_internet_service,     
df = X_train_t
#    one-hot encoding
onehot_cols = ["Sex", "HasMultiplePhoneNumbers", "HasInternetService",
               "HasOnlineSecurityService", "HasOnlineBackup", "HasDeviceProtection",
               "HasTechSupportAccess", "HasOnlineTV", "HasMovieSubscription",
               "PaymentMethod", "HasPhoneService"]


columns_to_encode = [
    "HasOnlineSecurityService",
    "HasOnlineBackup",
    "HasDeviceProtection",
    "HasTechSupportAccess",
    "HasOnlineTV",
    "HasMovieSubscription",
    'HasMultiplePhoneNumbers',
]

columns_to_encode2 = ['Sex',
 'PaymentMethod']

columns = ['HasInternetService', 'HasPhoneService']

categories = [["No", "No internet service", "Yes"] for _ in columns_to_encode]

preprocessor = ColumnTransformer(
    transformers=[
        #  service_cols:  "No"
        ("service_encoder", OneHotEncoder(
            categories=categories,
            drop=["No", "No", "No", "No", "No", "No", "No"],
            handle_unknown="ignore"
        ), columns_to_encode),

        #  binary_cols:   
        ("binary_encoder", OneHotEncoder(
            drop="first",
            handle_unknown="ignore"
        ), columns_to_encode2),

        #   :  OneHotEncoding
        ("other_encoder", OneHotEncoder(
            handle_unknown="ignore"
        ), columns)
    ],
    remainder="passthrough"
)


encoded_data = preprocessor.fit_transform(df)

feature_names = preprocessor.get_feature_names_out()

#   (onehot__, remainder__)
clean_feature_names = [name.split("__")[-1] for name in feature_names]

#   DataFrame
encoded_df = pd.DataFrame(encoded_data, columns=clean_feature_names)



#    "No internet service"
no_internet_cols = [col for col in encoded_df.columns if "No internet service" in col or "No phone service" in col]


#    "No internet service"
encoded_df.drop(columns=no_internet_cols, inplace=True)


#    label encoding
label_cols = [col for col in cat_cols if col not in onehot_cols]
X_train_encoded = encoded_df

if label_cols:
    for col in label_cols:
        le = LabelEncoder()
        X_train_encoded[col] = le.fit_transform(X_train_encoded[col])

X_train_encoded
```

```python
#     
#    preprocessor ( transform!)
X_test_encoded_data = preprocessor.transform(X_test_t)

#  DataFrame    
X_test_encoded = pd.DataFrame(
    X_test_encoded_data,
    columns=clean_feature_names  #      
)

#     "No internet service"
X_test_encoded.drop(columns=no_internet_cols, inplace=True, errors='ignore')

#  LabelEncoders    
if label_cols:
    #        (    train)
    #    , :
    label_encoders = {}
    for col in label_cols:
        le = LabelEncoder()
        X_train_encoded[col] = le.fit_transform(X_train_encoded[col])
        label_encoders[col] = le  #   

    #      
    for col in label_cols:
        le = label_encoders[col]
        #        
        X_test_encoded[col] = X_test_encoded[col].map(
            lambda s: le.transform([s])[0] if s in le.classes_ else -1
        )
```

```python
import seaborn as sns

#  X_train_t  y_train  
data = pd.concat([X_train_encoded, y_train], axis=1)

#   
corr_matrix = data.corr()

#   
plt.figure(figsize=(24, 20))

#   
sns.heatmap(
    corr_matrix,  #  
    annot=True,   #    
    fmt=".2f",    #       
    cmap="coolwarm",  #  
    vmin=-1,      #     
    vmax=1,       #     
    linewidths=0.5,  #    
    square=True   #   
)

#  
plt.title("  ", fontsize=16)

#  
plt.show()
```

```python
##  
log_reg = LogisticRegression(max_iter=1000)
reg = GridSearchCV(log_reg, param_grid={'C': [150, 130, 100, 50, 25, 10, 5, 1, 0.5, 0.25, 0.1, 0.05, 0.01, 0.001]},
                   scoring='roc_auc', refit=True, cv=7).fit(X_train_t, y_train)
best_model = reg.best_estimator_
```
