# Метрика

Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

```python
# разберемся с численными признаками
fig, axes = plt.subplots(nrows=8, ncols=5, figsize=(30, 20))
fig.suptitle('Визуализация данных', fontsize=16)
counter = 0
i = 0

for j, col in enumerate(num_cols):
    counter = j // 5
    i = j % 5
    # Гистограмма
    axes[counter, i].hist(data[col].dropna(), bins=5, color='skyblue', edgecolor='black')
    axes[counter, i].set_title(f'Гистограмма {col}')

    # Boxplot
    axes[counter + 1, i].boxplot(data[col].dropna(), vert=False, patch_artist=True, boxprops=dict(facecolor='lightgreen'))
    axes[counter + 1, i].set_title(f'Boxplot {col}')

old_counter = counter
old_i = i + 1

# Построение bar-диаграмм и круговых диаграмм для категориальных признаков
for j, col in enumerate(cat_cols):
    print(j, col)
    counter = old_counter + ((old_i + j) // 5) * 2
    i = (old_i + j % 5) % 5

    # Подсчёт количества значений
    value_counts = data[col].value_counts()

    # Bar-диаграмма
    axes[counter, i].bar(value_counts.index, value_counts.values, color='orange')
    axes[counter, i].set_title(f'Bar {col}')

    # Круговая диаграмма
    axes[counter + 1, i].pie(value_counts.values, labels=value_counts.index, autopct='%1.1f%%', colors=['lightcoral', 'lightblue'])
    axes[counter + 1, i].set_title(f'Pie {col}')

# Убираем лишние оси
for ax in axes.flatten():
    ax.grid(True)

plt.tight_layout()
plt.show()
```
